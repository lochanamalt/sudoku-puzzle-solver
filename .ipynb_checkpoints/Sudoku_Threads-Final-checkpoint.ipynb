```python
import time
import math
import numpy as np
import threading
import concurrent.futures
```

```python
emptycellsParallel=[]
emptycellsSerial=[]
data = np.loadtxt("puzzle.txt", dtype=np.int16)
dataSerial = data
```

```python
#finding empty cells and append them to the array
def findEmptyCellsParallel(data,e) :
    for row in range(3): 
        for col in range(9): 
            if(data[e + row][col]==0): 
                emptycellsParallel.append([e +row,col])

#finding empty cells and append them to the array
def findEmptyCellsSerial(data) :
    for row in range(9):
        for col in range(9): 
            if(data[row][col]==0): 
                emptycellsSerial.append([row,col])
                
#print grid
def grid(arr): 
    for i in range(9): 
        for j in range(9): 
            print(arr[i][j], end = ' ')
        print() 
        
def check_possibility_in_row(arr,row,num):
    for i in range(9):
        if(arr[row][i] == num):
            return False
    return True

def check_possibility_in_col(arr,col,num):
    for i in range(9):
        if(arr[i][col] == num):
            return False
    return True
        
def check_possibility_in_subsquare(arr,row,col,num):
    if((row+1)%3== 0):
        row = row - 2
    else:
        row = row + 1 - (row+1)%3
        
    if((col+1)%3== 0):
        col = col - 2  
    else:
        col = col + 1 - (col+1)%3  
        
    for i in range(3): 
        for j in range(3): 
            if(arr[i+row][j+col] == num):
                return False
    return True

def fillCells(data,i,filled,start_value,n):
    
    while(i<n):
        row = emptycellsParallel[i][0]
        col = emptycellsParallel[i][1] 

        for num in range(start_value,10):

            if(check_possibility_in_subsquare(data,row,col,num) and check_possibility_in_col(data,col,num) and check_possibility_in_row(data,row,num)):
                data[row][col] = num
                start_value = 1
                filled = 1
                break

        if(filled == 0):
            i = i-1
            row = emptycellsParallel[i][0]
            col = emptycellsParallel[i][1]        
            start_value = data[row][col] + 1
            data[row][col] = 0
        else:
            i = i+1
            filled = 0
            
```

```python
i=0
start_value = 1
filled = 0

start_time = time.time()

findEmptyCellsSerial(dataSerial)

print("Finding empty cells Serial--- %s seconds ---" % (time.time() - start_time))

start_time = time.time()

while(i<len(emptycellsSerial)):
    row = emptycellsSerial[i][0]
    col = emptycellsSerial[i][1] 
    for num in range(start_value,10):
        if(check_possibility_in_subsquare(dataSerial,row,col,num) and check_possibility_in_col(dataSerial,col,num) and check_possibility_in_row(dataSerial,row,num)):
            dataSerial[row][col] = num
            start_value = 1
            filled = 1
            break

    if(filled == 0):
        i = i-1
        row = emptycellsSerial[i][0]
        col = emptycellsSerial[i][1]        
        start_value = dataSerial[row][col] + 1
        dataSerial[row][col] = 0
    else:
        i = i+1
        filled = 0
        
grid(dataSerial)

print("--- Serial Prgramme %s seconds ---" % (time.time() - start_time))
```
